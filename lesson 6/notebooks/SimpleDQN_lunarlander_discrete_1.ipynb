```python
import os
import sys
import time
path_head = os.path.abspath('../../')
sys.path.append(path_head)
print(path_head)
import gymnasium as gym
from tqdm import tqdm
from rlcollection.simpledqn import RLDQN
from rlcollection.rlagents import DQNAgent
from IPython.display import HTML
from IPython.display import display
```

# One Agent. 2000 episodes

## target_net weights updating every 5 frames

```python
to_learn = 2_000  # episodes
env = gym.make("LunarLander-v2",
                render_mode=None,
                continuous=False,
                gravity=-9.8,
                enable_wind=True,
                wind_power=15.0,
                turbulence_power=1.5,
                )

rl = RLDQN(env, DQNAgent, agents_num=1, agents_devices=['cuda',])
rl.learn(to_learn, condition='episode')
rl.save(f'./simpledqqn_1_{to_learn}.pth')
```

```python
rl.learning_curve()
```

```python
rl.reset()
```

```python
env = gym.make("LunarLander-v2",
                render_mode='rgb_array_list',
                continuous=False,
                gravity=-9.8,
                enable_wind=True,
                wind_power=15.0,
                turbulence_power=1.5,
                )
```

```python
rl.set_env(env)
```

```python
gif_path_filename = f'./simpledqqn_1_{to_learn}.gif'
rl.save_movie_gif(gif_path_filename=gif_path_filename, 
                  weights_path_filename=f'./simpledqqn_1_{to_learn}.pth')
```

```python
_html_code = f'<img src="{gif_path_filename}">'
HTML(_html_code)
```

```python
rl.reset()
```

# Three Agents. 2000 episodes

 ## Updating weights with TAU between agents. Agents weights updating ~ every 19 frames, target_net weights updating every 5 frames

```python
to_learn = 2_000
env = gym.make("LunarLander-v2",
                render_mode=None,
                continuous=False,
                gravity=-9.8,
                enable_wind=True,
                wind_power=15.0,
                turbulence_power=1.5,
                )

rl = RLDQN(env, DQNAgent, agents_num=3, agents_devices=['cuda', 'cpu', 'cpu'])
rl.learn(to_learn, condition='episode')
rl.save(f'./simpledqqn_3_{to_learn}.pth')
```
